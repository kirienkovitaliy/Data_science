### # Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
tables = pd.read_html(url)
df = tables[12]
print(df)
```

### # Вивести перші рядки таблиці за допомогою методу head

```python
print(df.head())
```

### # Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
print(df.shape)
```

### # Замініть у таблиці значення "—" на значення NaN

```python
df = df.replace(to_replace = "—", value = np.NaN)
print(df)
```

### # Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
print(df.dtypes)
```

### # Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
df = df.apply(pd.to_numeric, errors = "ignore")
print(df.dtypes)
```

### # Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
print(df.isnull().sum() / len(df) * 100)
```

### # Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
df = df.drop([27])
print(df)
```

### # Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
df = df.fillna(df.mean(numeric_only=True)).round(1)
print(df)
```

### # Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
print(df[df["2019"] > df["2019"].mean()]["Регіон"].tolist())
```

### # У якому регіоні була найвища народжуваність у 2014 році?

```python
print(df[df["2014"] == df["2014"].max()]["Регіон"].iloc[0])
```

### # Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.bar(df["Регіон"], df["2019"])
plt.xticks(rotation=45, ha="right")
plt.xlabel("Регіон")
plt.ylabel("Коефіцієнт народжуваності")
plt.title("Народжуваність по регіонах України у 2019 році")
plt.tight_layout()

plt.show()



```

```python
%matplotlib inline
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['2019'], bins=15, kde=True)
plt.xlabel('Рівень народжуваності')
plt.ylabel('Кількість регіонів')
plt.title('Гістограма рівня народжуваності у 2019 році')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['2000'], y=df['2019'], color='b', alpha=0.7)
plt.xlabel('Рівень народжуваності у 2000 році')
plt.ylabel('Рівень народжуваності у 2019 році')
plt.title('Залежність рівня народжуваності в 2019 році від рівня в 2000 році')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop('Регіон', axis=1))
plt.xticks(rotation=90)
plt.title('Розподіл рівня народжуваності за всі роки')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
selected_regions = ['Київ', 'Львівська', 'Харківська', 'Одеська', 'Черкаська']
for region in selected_regions:
    plt.plot(df.columns[1:], df[df['Регіон'] == region].iloc[0, 1:], label=region)

plt.xlabel('Рік')
plt.ylabel('Рівень народжуваності')
plt.title('Динаміка рівня народжуваності за роками для обраних областей')
plt.legend()
plt.grid(True)
plt.show()
```
